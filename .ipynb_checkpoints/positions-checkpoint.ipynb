## Gaussian fit - counts and error

```python
def model(x, A, mu, sigma, B, c):
    return (A / (np.sqrt(2*np.pi)*sigma)) * np.exp(-(x-mu)**2/(2*sigma**2)) + B*x + c
```

### POSITION 1
most left, middle hole (4,1)(row, column): 4.2cm from LaBr3:Ce, 10.2 cm from CeBr

```python
#initial guesses 
A01 = np.sum(y1)
mu01 = x1[np.argmax(y1)]
sigma01 = 50
B01 = 0
c01 = np.min(y1)

p01 = [A01, mu01, sigma01, B01, c01]
```

```python
popt1, pcov1 = curve_fit(model, x1, y1, p0=p01)

A1, mu1, sigma1, B1, c1 = popt1
A1_err = np.sqrt(pcov1[0,0])

print(f"Peak channel = {mu1:.2f}")
print(f"Counts (A) = {A1:.2f} ± {A1_err:.2f}")
```

```python
plt.scatter(x1, y1, s=10, label="Data")
plt.plot(x1, model(x1, *popt1), 'r-', label="Fit")
plt.xlabel("ADC Channel")
plt.ylabel("Counts")
plt.legend()
plt.show()
```

## current activity 

```python
A0 = 36200  # initial activity [Bq]
half_life_years = 2.6
half_life_seconds = half_life_years * 365.25 * 24 * 3600

date_initial = datetime(2010, 6, 1)
date_measured = datetime(2025, 12, 17)

delta_t = (date_measured - date_initial).total_seconds()

current_activity = A0 * np.exp(-np.log(2) * delta_t / half_life_seconds)

print(f"Current activity (Bq): {current_activity:.2f}")
```

## geometric efficiency (%)

```python
branching_ratio = 179.91 / 100  # gamma emission probability

efficiency1 = (A1 / (current_activity * t1 * branching_ratio)) * 100

print(f"Geometric efficiency: {efficiency1:.4f} %")
```

```python
# relative uncertainties
rel_counts1 = A1_err / A1
rel_activity1 = 0.10

# total relative uncertainty
rel_total1 = np.sqrt(rel_counts1**2 + rel_activity1**2)

efficiency_error1 = efficiency1 * rel_total1

print("Efficiency uncertainty (%):", efficiency_error1)
```
